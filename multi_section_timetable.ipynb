#Clean btech

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('btech.csv')
df.head()
```

```python
df = df.drop(['Unnamed: 0', '#S', 'L T P C/\nL P U'], axis=1)
df = df.drop(21, axis=0)
df.reset_index(inplace=True, drop=True)
df.tail()
```

```python
for i in range(len(df) - 1):
    if df['Instructor Name'].iloc[i] == df['Instructor Name'].iloc[i+1]:
        if df['Title'].iloc[i+1] == 'LAB' or df['Code'].iloc[i+1] == 'LAB':
            print(i,"and",i+1)
            
            for col in df.columns[:6]:
                if pd.isna(df[col].iloc[i+1]):        
                    df[col].iloc[i+1] = df[col].iloc[i]
            df['Title'].iloc[i+1] = '[LAB] '+df['Title'].iloc[i]
```

```python
df['Day1'] = df['Day'].str.split(',')
df['Session1'] = df['Session'].str.split(',')
```

```python
df = df.drop(['Day', 'Session'], axis=1)
```

```python
df['Session1'] = df['Session1'].apply(lambda x: [int(item) for item in x])
df
```

```python
df['Day1'] = df.apply( lambda x: x['Day1']*len(x['Session1']) if ( len(x['Day1']) != len(x['Session1']) ) else x['Day1'], axis=1)
```

```python
df[['Day1', 'Session1']]
```

```python
df.columns
```

```python
df = df.explode(['Day1', 'Session1'])
df = df.rename(columns={'Day1':'Day', 'Session1':'Session'})
df = df[['Title','Sections', 'Day','Session', 'Common Hour','Room', 'Test1', 'Test2', 'Compre','Instructor Name']]
df.reset_index(drop=True, inplace=True)
```

```python
df['Title'] = df['Title'].str.replace("Fundamentals of Signal Processing", 'Signal Processing')
df['Title'] = df['Title'].str.replace("Computer organization and Architecture", 'Computer Org')
df['Title'] = df['Title'].str.replace("Web Enabled Technologies", 'Web Tech')
df['Title'] = df['Title'].str.replace("Dynamics Of Social Change", 'Social Change')
df['Title'] = df['Title'].str.replace("Database Management Systems", 'Database Management')

df['Day'] = df['Day'].str.replace("M", 'Monday')
df['Day'] = df['Day'].str.replace("TU", 'Tuesday')
df['Day'] = df['Day'].str.replace("W", 'Wednesday')
df['Day'] = df['Day'].str.replace("TH", 'Thursday')
df['Day'] = df['Day'].str.replace("F", 'Friday')
```

```python
df.head()
```

```python
df['Common Hour'] = df[df['Common Hour'].notnull()]['Common Hour'].str.extract(r'(\w+)\((\d+)\)').apply(lambda x: [x[0], int(x[1])], axis=1)
```

#Timetable

```python
section = input("Do u belong to one section only\n-> If so type your section---> A,B,C,D,1\nelse type '!N'  \t: ").upper()
if section == '!N':
    subjects = {v+1: k for v, k in enumerate(set(df['Title']))}
    for num, sub in subjects.items():
        print(f"{num}. \t: {sub}")
    print("remember to press q to exit\n")
    user={}
    lst=['A', 'B', 'C', 'D', '1']
    while True:

        inp = input(f"\n->Choose your subject by entering the serial number: ")
        if inp == 'q':
            break
        subject = subjects[int(inp)]
        if subject in user.values():
            print(f"{subject} is already registered, do u want to override [Y/n]")
            if  input(": ").lower() == 'n':
                print('ignoring')
                continue
        print('Choose from ', lst)
        section = input(f"Now, section for {subject}: ").upper()
        if section not in lst:
            print('section not found, rewinding time')
            print("...")
            continue
        if section != '1':
            section= 'Section-'+section
        
        user[subject] = section

    section_values = list(user.values())
    subject_values = list(user.keys())
    print(section_values)
    print(subject_values)
    sec = df[ df['Title'].isin(subject_values) & df['Sections'].isin(section_values) ]
else:
    sec = df[df['Sections'] == 'Section-'+section]
```

```python
sec.head()
```

```python
print("Classes per week:",len(sec))
def per_day(day):
    week_day = sec[sec['Day'] == day]
    return week_day.set_index('Session')['Title'].to_dict()
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
timetable={day : per_day(day) for day in week}

sec = sec[sec['Common Hour'].notnull()]
ch = sec[['Title', 'Common Hour']].set_index('Title').to_dict().get('Common Hour')

def add_commn_hr(timetable, subject, day, period):
    if day not in timetable:
        timetable[day]={}
    timetable[day][period] = subject
    return timetable

for subject, (day, period) in ch.items():
    timetable = add_commn_hr(timetable, subject+" `C~Hr", day, period)

tb = pd.DataFrame(timetable)
tb = tb.sort_index(axis=0)
if input('Do you require the periods to be horizontal [Y/n]: ').lower() == 'y':
    tb=tb.T
tb
```

```python
if section != '!N':
    tb.to_csv('Section-'+section+'.csv')
else:
    inp = input("Filename to save?: ")
    tb.to_csv(inp+'.csv')
```
